```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 20000

campaigns = [f"Campaign {i}" for i in range(1, 9)]
ad_sets = [f"Ad Set {i}" for i in range(1, 20)]
ads = [f"Ad Creative {i}" for i in range(1, 40)]

ages = ["18-24","25-34","35-44","45-54","55+"]
genders = ["Male","Female"]

dates = pd.date_range("2024-01-01","2024-06-30")

df = pd.DataFrame({
    "Campaign Name": np.random.choice(campaigns, rows),
    "Ad Set Name": np.random.choice(ad_sets, rows),
    "Ad Name": np.random.choice(ads, rows),
    "Day": np.random.choice(dates, rows),
    "Age": np.random.choice(ages, rows),
    "Gender": np.random.choice(genders, rows)
})
```

```python
# Base performance tiers for campaigns
campaign_strength = {
    "Campaign 1": 1.3,
    "Campaign 2": 1.2,
    "Campaign 3": 1.1,
    "Campaign 4": 1.0,
    "Campaign 5": 0.9,
    "Campaign 6": 0.85,
    "Campaign 7": 0.8,
    "Campaign 8": 0.75
}

# Audience engagement modifiers
audience_boost = []

for age, gender in zip(df["Age"], df["Gender"]):
    
    boost = 1
    
    if age == "25-34":
        boost *= 1.25
        
    if gender == "Female":
        boost *= 1.1
        
    audience_boost.append(boost)

df["audience_factor"] = audience_boost
```

```python
# Generate impressions
df["Impressions"] = np.random.randint(1000,20000,rows)

# Reach and frequency
reach_factor = np.random.uniform(0.6,0.9,rows)
df["Reach"] = (df["Impressions"] * reach_factor).astype(int)
df["Frequency"] = df["Impressions"] / df["Reach"]

# Base CTR
base_ctr = np.random.uniform(0.008,0.025,rows)

# Apply campaign strength
campaign_multiplier = df["Campaign Name"].map(campaign_strength)

# Apply audience factor
ctr = base_ctr * campaign_multiplier * df["audience_factor"]

# Simulate ad fatigue (CTR decreases if frequency is high)
ctr = ctr * (1 / (1 + (df["Frequency"] - 1)*0.4))

df["CTR (All)"] = ctr.clip(0.003,0.05)
```

```python
# Clicks
df["Clicks (All)"] = (df["Impressions"] * df["CTR (All)"]).astype(int)

# CPC varies by campaign
df["CPC (All)"] = np.random.uniform(0.2,2.0,rows) / campaign_multiplier

# Spend
df["Amount Spent (PHP)"] = df["Clicks (All)"] * df["CPC (All)"]

# CPM
df["CPM (Cost per 1,000 Impressions)"] = (
    df["Amount Spent (PHP)"] / df["Impressions"]
) * 1000

# Engagement correlated with clicks
df["Post Engagement"] = (
    df["Clicks (All)"] * np.random.uniform(1.5,3.0,rows)
).astype(int)

df["Currency"] = "PHP"

df["Objective"] = np.random.choice(
    ["Traffic","Conversions","Engagement","Brand Awareness"],rows
)

df["call_to_action_type"] = np.random.choice(
    ["Shop Now","Learn More","Sign Up","Download"],rows
)

df["Reporting Starts"] = df["Day"]
df["Reporting Ends"] = df["Day"]

# Remove helper column
df = df.drop(columns=["audience_factor"])
```

```python
# Place resulting dataset in data directory
df.to_csv("../data/synthetic_ads_data.csv", index=False)

print("Advanced synthetic dataset created!")
```
